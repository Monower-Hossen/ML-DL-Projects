````python
# Project 2: Diabetes Prediction using Machine Learning (SVM)

## 1. Project Overview

The goal of this project is to build an **end-to-end machine learning pipeline** in Python to predict whether a person has diabetes based on medical attributes. We use a **Support Vector Machine (SVM)** classifier.

This is a **binary classification problem**:

* `0` → Non-diabetic
* `1` → Diabetic

---

## 2. Dataset

We use the **Pima Indians Diabetes Dataset**, which contains medical predictor variables and one target variable.

**Features:**

* Pregnancies
* Glucose
* BloodPressure
* SkinThickness
* Insulin
* BMI
* DiabetesPedigreeFunction
* Age

**Target:**

* Outcome

---

## 3. Tools & Libraries

* Python
* NumPy
* Pandas
* Matplotlib / Seaborn
* Scikit-learn

---

## 4. Workflow (End-to-End)

1. Import Libraries
2. Load Dataset
3. Data Exploration & Cleaning
4. Feature Scaling
5. Train-Test Split
6. Model Training (SVM)
7. Model Evaluation
8. Prediction on New Data

---

## 5. Implementation (Python Code)



```python
```

---

### 5.2 Load Dataset

```python

```

---

### 5.3 Data Exploration

```python
print(diabetes.shape)
print(diabetes.info())
print(diabetes.describe())
```

Check class distribution:

```python
diabetes['Outcome'].value_counts()
```

---

### 5.4 Feature Selection

```python
X = diabetes.drop('Outcome', axis=1)
y = diabetes['Outcome']
```

---

### 5.5 Feature Scaling

SVM performs better when features are scaled.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

---

### 5.6 Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

---

### 5.7 Model Training (Support Vector Machine)

```python
svm_model = SVC(kernel='linear', C=1.0)
svm_model.fit(X_train, y_train)
```

---

### 5.8 Model Evaluation

#### Accuracy

```python
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

#### Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

#### Classification Report

```python
print(classification_report(y_test, y_pred))
```

---

### 5.9 Predicting on New Data

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

input_array = np.asarray(input_data)
input_reshaped = input_array.reshape(1, -1)
input_scaled = scaler.transform(input_reshaped)

prediction = svm_model.predict(input_scaled)

if prediction[0] == 0:
    print('The person is NOT diabetic')
else:
    print('The person is diabetic')
```

---

## 6. Results

* The SVM model achieves **~75–80% accuracy** (may vary by split and parameters).
* Feature scaling significantly improves performance.

---

## 7. Conclusion

* Support Vector Machine is effective for diabetes prediction.
* Standardization is crucial for SVM.
* The model can be improved using:

  * Hyperparameter tuning (GridSearchCV)
  * Different kernels (RBF, Poly)
  * Handling zero or missing values properly

---

## 8. Future Enhancements

* Build a **Flask / Streamlit web app**
* Compare with Logistic Regression, Random Forest
* Deploy model using Pickle

---

## 9. Final Note

This project demonstrates a **complete ML pipeline** suitable for:

* Academic projects
* Portfolio / resume
* Beginner-to-intermediate ML learning

---

✅ *End of Project*
````

### 1. Import Libraries

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
````

### 2.Load Dataset

````python
diabetes = pd.read_csv('diabetes.csv')
diabetes.head()
````

### 3. Data Exploration

````python
diabetes.shape
````

````python
diabetes.info()
````

````python
diabetes.describe()
````

### 4. Feature Selection

````python
diabetes['Outcome'].value_counts()
````

### 5.Feature Selection

````python
X = diabetes.drop('Outcome', axis=1)
y = diabetes['Outcome']
````

````python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

````python
X_train, X_test, y_train, y_test = train_test_split(
X_scaled, y, test_size=0.2, random_state=42
)
````

````python
svm_model = SVC(kernel='linear', C=1.0)
svm_model.fit(X_train, y_train)
````

````python
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
````

````python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
````

````python
print(classification_report(y_test, y_pred))
````

````python
input_data = (5,166,72,19,175,25.8,0.587,51)


input_array = np.asarray(input_data)
input_reshaped = input_array.reshape(1, -1)
input_scaled = scaler.transform(input_reshaped)


prediction = svm_model.predict(input_scaled)


if prediction[0] == 0:
print('The person is NOT diabetic')
else:
print('The person is diabetic')
````
