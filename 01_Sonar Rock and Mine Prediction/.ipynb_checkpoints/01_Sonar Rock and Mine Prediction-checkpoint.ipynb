# Project 01 :Sonar Rock and Mine Prediction

## Project Overview

This project focuses on building a **machine learning classification model** that can distinguish between **rocks (R)** and **mines (M)** using **sonar signal data** collected in underwater environments. Accurate detection is critical for naval safety, underwater exploration, and autonomous underwater vehicles (AUVs).

## Problem Statement

Underwater sonar systems emit sound waves and analyze the reflected signals. However, rocks and mines often produce **similar sonar returns**, making manual or rule-based detection unreliable.
The goal is to **automatically classify sonar returns** as either *rock* or *mine* using supervised learning.

## Dataset

* **Commonly used dataset:** UCI Sonar Dataset
* **Features:**

  * 60 numerical attributes
  * Each represents energy within a specific frequency band
* **Target classes:**

  * `R` → Rock
  * `M` → Mine

## Machine Learning Approach

### 1. Data Preprocessing

* Handle missing values (if any)
* Feature scaling (StandardScaler / MinMaxScaler)
* Train-test split (e.g., 80% training, 20% testing)

### 2. Models Used

Typical models for this project:

* Logistic Regression
* Support Vector Machine (SVM)
* Random Forest Classifier
* K-Nearest Neighbors (KNN)
* Neural Networks (optional advanced approach)

## Testing & Evaluation Strategy

### Evaluation Metrics

* **Accuracy**
* **Precision**
* **Recall**
* **F1-score**
* **Confusion Matrix**

### Testing Methodology

* Test model on unseen sonar signals
* Use **cross-validation** for robustness
* Compare performance across multiple models

## Expected Results

* SVM and Random Forest often achieve **high accuracy**
* Feature scaling significantly improves performance
* False negatives (mines classified as rocks) are treated as **critical errors**

## Applications

* Naval mine detection
* Autonomous underwater vehicles (AUVs)
* Submarine navigation safety
* Marine research and exploration

## Conclusion

The project demonstrates how machine learning can significantly improve **underwater object classification** using sonar data. With proper preprocessing and model selection, reliable mine detection can be achieved even in noisy underwater environments.

```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import joblib  # For saving the trained model
```

# 1 Get Data

```python
solar_data = pd.read_csv('sonar.csv',header=None)
solar_data.head()
```

# 2 Preprocess Data

```python
solar_data.shape
```

```python
solar_data.duplicated().sum()
```

```python
solar_data[60].value_counts()
```

```python
solar_data.groupby(60).mean()
```

```python
solar_data.describe()
```

# 3 Train Test Split

```python
X = solar_data.drop(columns=60,axis=1)
y = solar_data[60]
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, stratify=y, random_state=1)
```

```python
print(X_train.shape,X_test.shape,'\n')
print(y_train.shape,y_test.shape)
```

```python
# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# 4 Build and Evualivate Model

```python
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale')  # RBF kernel
svm_model.fit(X_train_scaled, y_train)
```

```python
y_pred = svm_model.predict(X_test_scaled)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Rock','Mine'], yticklabels=['Rock','Mine'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
